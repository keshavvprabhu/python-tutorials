## Trick 7: Convert Strings to Numbers

```python
price_df = pd.DataFrame({'AsofDate': ['2022-05-23', '2022-05-24', '2022-05-25'], 
                   'ClosePrice':['123.45', '145.56', '120.98'],
                   'AdjustedClosePrice': ['0.01', '0.03', 'NA']
                   })
price_df
```

```python
price_df.dtypes
```

```python
pd.to_numeric(price_df.ClosePrice)
```

```python
# pd.to_numeric(price_df.AdjustedClosePrice)

# ValueError                                Traceback (most recent call last)
# /usr/local/lib/python3.7/dist-packages/pandas/_libs/lib.pyx in pandas._libs.lib.maybe_convert_numeric()

# ValueError: Unable to parse string "NA"
```

```python
pd.to_numeric(price_df.AdjustedClosePrice, errors='coerce')
```

If you know for sure that NaN can be represented as Zeros then you can use the fillna(0) function. This function will replace all NaN with 0

```python
pd.to_numeric(price_df.AdjustedClosePrice, errors='coerce').fillna(0)
```

If you want to apply this rule to the entire dataframe then do the following:

```python
converted_price_df = price_df[['ClosePrice', 'AdjustedClosePrice']].apply(pd.to_numeric, errors='coerce').fillna(0)
```

But note that the changes we applied did not impact the original price_df

```python
price_df
```

Now to apply the change to the price_df, we need to take a column by column approach, because otherwise, all the columns in the dataframe could be converted to integers forcefully, which may not be something that we want

```python
price_df['ClosePrice'] = pd.to_numeric(price_df.ClosePrice, errors='coerce')
price_df['AdjustedClosePrice'] = pd.to_numeric(price_df.AdjustedClosePrice, errors='coerce')
```

```python
price_df
```

## Trick 8: Reduce Dataframe Size

```python
drinks_df = pd.read_csv('http://bit.ly/drinksbycountry')
drinks_df
```

```python
drinks_df.info(memory_usage='deep')
```

Read only the columns that you need by using usecols argument

```python
cols = ['beer_servings', 'continent']
beer_by_continent_df = pd.read_csv('http://bit.ly/drinksbycountry', usecols=cols)
beer_by_continent_df.info(memory_usage='deep')
```

Change the datatypes from object to category if the number of different values are small

```python
drinks_df.continent.value_counts()
```

In this case there are only 6 continents in the dataset. Antartica does not serve drinks! 😀

As such, it is a good idea to convert the datatype of the continent column to category

```python
dtypes_changes = {'continent': 'category'}
cols = ['beer_servings', 'continent']
smaller_beer_by_continent_df = pd.read_csv('http://bit.ly/drinksbycountry', usecols=cols, dtype=dtypes_changes)
smaller_beer_by_continent_df.info(memory_usage='deep')
```
