```python
unrate = fred.get_series('UNRATE')
unrate.plot()
```

```python
unemp_state = fred.search("unemployment state", filter=('frequency', 'Monthly'))
unemp_state
```

```python
unemp_state_df = unemp_state.query('seasonal_adjustment == "Seasonally Adjusted" and units == "Percent"')
```

```python
unemp_state_df.loc[unemp_state_df['title'].str.contains('Unemployment Rate')]
unemp_state_df
```

```python
unemp_state_df.index
```

```python
len(unemp_state_df.index)
```

```python
all_results = []
for myid in unemp_state_df.index:
    results = fred.get_series(myid)
    results = results.to_frame(name=myid)
    all_results.append(results)

final_df = pd.concat(all_results, axis=1)
```

```python
t = [col for col in final_df.columns if len(col) <= 4]
df_unemp_states = final_df[t].dropna()
```

```python
px.line(df_unemp_states)
```
