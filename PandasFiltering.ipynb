# How do I filter data using multiple conditions

```python
import pandas as pd

df = pd.read_csv('http://bit.ly/imdbratings')
df.head()
```

```python
df[(df['duration'] > 200) & (df['genre'].isin(['Drama', 'Crime']))]
```
